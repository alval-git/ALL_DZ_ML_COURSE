```python
#импорты
from data_utils import preprocess_data, load_data, get_df_info
from sklearn.model_selection import train_test_split
from model_utils import train_model, predict, evaluate_model
from plot_utils import real_and_predict_target_plot
```

```python
#подготовка 
df = load_data("./winequality-white.csv", ';')
get_df_info(df)
numeric_params = ["fixed acidity", "volatile acidity", "citric acid", "residual sugar", "chlorides", "free sulfur dioxide",	"density",	"pH", "sulphates", "quality"]
target = "alcohol"
X, y, preprosessor = preprocess_data(df, target, numeric_params)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=20, shuffle=False)
```

```python
#запуск 
model = train_model(X_train, y_train)
y_pred = predict(model, X_test)
mse, r2 = evaluate_model(y_test, y_pred)
print(f'mse - {mse}, r2 - {r2}')
real_and_predict_target_plot(y_pred, y_test, target)
```
